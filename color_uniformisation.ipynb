```python
import cv2
import numpy as np
import os

train_folder=os.listdir("data/resizedTest")

for folder in train_folder: 
    if folder != '.DS_Store': 
        images=os.listdir("data/resizedTest/"+folder)
        for image in images :
            if image != '.DS_Store' : 
                print('data/resizedTest/'+folder+'/'+image)
                img = cv2.imread('data/resizedTest/'+folder+'/'+image, cv2.IMREAD_GRAYSCALE)
                if np.sum(img == 255)>np.sum(img == 0): 
                    img = cv2.bitwise_not(img)
                    
                cv2.imwrite(os.path.join('data/colorsizeTest/'+folder+'/' , image), img)
```
