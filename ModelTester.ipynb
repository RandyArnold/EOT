```python
for class_folder in train_folder: 
    if class_folder != '.DS_Store': 
        images=os.listdir(train_folder_name+class_folder)
        for image in images :
            if image != '.DS_Store' : 
                #print(class_folder+'/'+image)
                img = keras.preprocessing.image.load_img(
                    train_folder_name+class_folder+'/'+image, target_size=(img_height, img_width)
                )
                img_array = keras.preprocessing.image.img_to_array(img)
                img_array = tf.expand_dims(img_array, 0) # Create a batch
                
                predictions = model.predict(img_array)
                scores = tf.nn.softmax(predictions[0])
                
                results[str(class_folder)][str(class_names[np.argmax(scores)])]+=1
                
                #print('{:.2f}'.format(score[1]*100) )
 
                #print(
                #    "This image most likely belongs to class {} with a {:.2f} percent confidence."
                #    .format(class_names[np.argmax(scores)], 100 * np.max(scores)))
            
```

```python
# SINGLE TRAIN WITH SCORES
image='data/testDataSet/6/86-108_3.png'

img = keras.preprocessing.image.load_img(
    image, target_size=(img_height, img_width)
)
img_array = keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = model.predict(img_array)
scores = tf.nn.softmax(predictions[0])

 
print(
    "This image most likely belongs to class {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(scores)], 100 * np.max(scores)))

print('Scores : ')

for i in range(24): 
    print("Score for class {} : {:.2f}".format(class_names[i], scores[i]*100))
```
